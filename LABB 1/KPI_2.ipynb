```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("../Data/Folkhalsomyndigheten_Covid19_Vaccine.xlsx", sheet_name= "Vaccinerade kön")
df
```

## Men doses

```python
men_1_dose = df.loc[2:2, "Antal vaccinerade"].sum()
men_2_dose = df.loc[3:3, "Antal vaccinerade"].sum()
print(f"Men who took atleast one dose: {men_1_dose}")
print(f"Men who took atleast two doses: {men_2_dose}")
```

## Women doses

```python
women_1_dose = df.loc[4:4, "Antal vaccinerade"].sum()
women_2_doses = df.loc[5:5, "Antal vaccinerade"].sum()
print(f"Women who took atleast one dose: {women_1_dose}")
print(f"Women who took atleast one dose: {women_2_doses}")
```

```python
total_1_dose = df.loc[:0, "Antal vaccinerade"].sum()
total_2_dose = df.loc[1:1, "Antal vaccinerade"].sum()
print(total_1_dose)
print(total_2_dose)
```

# Men 1 dose vs 2

```python
visual = np.array([men_1_dose, men_2_dose])

plt.pie(visual, labels = ["Men 1 dose", "Men 2 doses"], )
plt.legend(title = "Number of doses")
plt.show()
plt.savefig("Visualiseringar/men_1_vs_2.png")
```

# Women 1 dose vs 2

```python
visual2 = np.array([women_1_dose, women_2_doses])

plt.pie(visual2, labels = ["Women 1 dose", "Women 2 doses"])
plt.legend(title = "Number of doses")
plt.show()
plt.savefig("Visualiseringar/women_1_vs_2.png")
```

# Total 1 dose vs 2

```python
total = np.array([total_1_dose, total_2_dose])

plt.pie(total, labels = ["Total 1 dose","Total 2 doses"])
plt.legend(title = "Number of doses")
plt.show()
plt.savefig("Visualiseringar/total_one_dose_vs_2.png")
```

```python
percent_diff_total = (total_1_dose / total_2_dose)*100
percent_diff_total
```

```python
percent_diff_women = (women_1_dose / women_2_doses)*100
percent_diff_women
```

```python
percent_diff_men = (men_1_dose / men_2_dose)*100
percent_diff_men
```

```python
percent_diff_genders_1_dose = (men_1_dose / women_1_dose)*100
percent_diff_genders_1_dose
```

```python
men_women_diff = np.array([men_1_dose, women_1_dose])

plt.pie(men_women_diff, labels = ["Men 1 dose","Women 2 dose"])
plt.legend(title = "Number of doses")
plt.show()
plt.savefig("Visualiseringar/men_1_women_1.png")
```

* These KPI's are used to represent the data of number of doses across genders in Sweden.
* It's possible to determine that women have been vaccinated to a higher degree than men, especially considering that there are more men than women in sweden.
